```python
import numpy as np 
import pandas as pd  
import seaborn as sns 
```

```python
# 1 loading dataset, df 
df = pd.read_csv('titanic.csv')
df
```

```python
# 2 performing basic data exploration  
df.info()
```

```python
df.describe()
```

```python
# 3 handling missing values  
df.isnull().sum()
```

```python
# using median imputation to handle missing values in the Age Column

df['Age'] = df['Age'].fillna(df['Age'].mean())

# checking if null values still exist in Age column 

df.isnull().sum()
```

```python
 # using mode imputation to fill in missing values in Cabin Column 

df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])

# checking if missing values still exist in the Cabin Column 

df.isnull().sum()
```

```python
# Using mode imputation to fill in the missing value in the Embarked Column 

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0]) 

# Checking if the Embarked Column no longer has missing values 

df.isnull().sum()
```

Question 1: Create a countplot to show the number of passengers by gender

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
sns.countplot(
    x = 'Sex',
    data = df
)
plt.title('A count of passengers by gender') 
plt.xlabel('Sex')
plt.ylabel('Count of Passengers') 
plt.show()
```

Question 2: Create a bar chart to show the number of passengers in each passenger class using different colors.


```python
sns.countplot(
    x= df['Pclass'],
    data = df ,
    palette= 'Set2'
)
plt.title('Number of passengers in each passenger class') 
plt.xlabel('Passenger class') 
plt.ylabel('Number of passengers') 
plt.show()
```

Question 3: Create a histogram to show the distribution of passenger age.


```python
sns.histplot(df['Age'],bins=8)
plt.title('Passenger Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Question 4: Create a boxplot to detect outliers in the fare column.


```python
sns.boxplot(df['Fare'])
plt.title('Boxplot for Passenger Fare')
plt.show()
```

Question 5. Create a scatterplot to:

a. show the relationship between age and fare.

b. analyze the relationship between family size (SibSp + Parch) and fare.

```python
# a. the relationship between age and fare.
sns.scatterplot(x= df['Age'], y= df['Fare'])
plt.title('Age Vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
# b. relationship between family size (SibSp + Parch) and fare. 

df['Family size'] = df['SibSp'] + df['Parch']

sns.scatterplot(x = df['Family size'], y= df['Fare'])
plt.title('Family size vs Fare')
plt.xlabel('family size')
plt.ylabel('fare')
plt.grid(True)
plt.show()
```

Question 6: Create a countplot to show survival count by gender.

```python
sns.countplot(x=df['Sex'], hue=df['Survived'],data=df)
plt.title('Survival count by gender')
plt.xlabel('Gender')
plt.ylabel('Survival Count') 
plt.legend(title = 'Survived', labels=['No','Yes'])
plt.show()
```

Question 7: Create a bar chart to show survival by passenger class.

```python
sns.countplot(x=df['Pclass'], hue = df['Survived'])
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Count')
plt.legend(title = 'Survived', labels=['No','Yes'])
plt.show()
```

Question 8: Create a heatmap to show correlation between numerical variables.

```python
corr = df.corr(numeric_only=True) 
sns.heatmap(corr,annot=True, cmap='coolwarm',linewidths= 1.5)
plt.title('correlation heatmap')
plt.show()
```
